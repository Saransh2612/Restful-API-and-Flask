#1. What is a RESTful API?
--- A RESTful API (Representational State Transfer API) is a type of web API that follows a set of architectural principles defined by REST, a standard for designing networked applications. It allows different systems to communicate over the internet using standard HTTP methods.

#2. Explain the concept of API specification ?
--- An API specification is a formal, structured document or format that defines how a software component or system interacts with others through an Application Programming Interface (API). It outlines the expected endpoints, methods, input parameters, data formats, authentication methods, and responses involved in the communication.

#3. What is Flask, and why is it popular for building APIs?
--- Flask is a lightweight, open-source web framework written in Python. It is used to build web applications, especially RESTful APIs. Flask is based on the Werkzeug WSGI toolkit and Jinja2 templating engine.

Flask is known as a micro-framework because it provides the core functionality but leaves out many features that can be added as needed using extensions.

#4. What is routing in Flask?
--- In Flask, routing is the process of mapping URLs (web addresses) to specific functions in your Python code. These functions are called view functions, and they define what should happen when someone visits a certain URL in the web application or API.

#7. What is the purpose of the @app.route() decorator in Flask?
--- In Flask, the @app.route() decorator is used to bind a URL (route) to a function, also known as a view function.

It creates a mapping between URLs and Python functions.

This is how Flask knows what content to show when someone visits a specific web address in your app.



#8. What is the difference between GET and POST HTTP methods?
--- GET,	Retrieve data from the server
POST,	Send data to the server to create or process something

GET,	Data is sent in the URL as query parameters (e.g., ?name=John)
POST,	Data is sent in the request body (hidden from URL)

#10. How do you connect Flask to a SQL database?
--- To connect a Flask app to a SQL database, you typically use an ORM (Object-Relational Mapper) like SQLAlchemy. Flask provides great support for this through the Flask-SQLAlchemy extension.



#11. What is the role of Flask-SQLAlchemy?
--- Flask-SQLAlchemy is an extension for Flask that integrates the SQLAlchemy ORM (Object-Relational Mapper) with Flask applications, making it easier to work with relational databases.

Simplifies Database Integration
It provides a simple and Flask-friendly way to connect your app to SQL databases (like SQLite, MySQL, PostgreSQL) without writing raw SQL queries.

ORM Functionality
It allows you to define database tables as Python classes (models) and interact with database records as Python objects, improving readability and maintainability.

Manages Database Sessions
Flask-SQLAlchemy handles the creation and management of database sessions automatically, so you don’t have to manage database connections explicitly.

#18. What is an API specification, and how does it help in building a Flask API?

#23. What is the role of Flask’s session object?

```python
#1. How do you create a basic Flask application?
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello, World!'

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#2. How do you serve static files like images or CSS in Flask?
from flask import Flask, send_from_directory

app = Flask(__name__)

@app.route('/static/<path:filename>')
def serve_static(filename):
    return send_from_directory('static', filename)
```
